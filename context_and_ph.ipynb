```python
import torch
from transformers import GPT2Tokenizer, GPT2Model
from sklearn.decomposition import PCA
import plotly.graph_objects as go

# Initialize the model and tokenizer
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2Model.from_pretrained('gpt2')

sentences = ["Quantum information theory is interesting.", 
             "Quantum information theory allows us to study attention using entanglement."]

# Function to calculate Euclidean distance
def euclidean_distance(a, b):
    return torch.dist(a, b)

# Iterate over sentences
for sentence in sentences:
    # Tokenize the sentence
    inputs = tokenizer(sentence, return_tensors='pt')

    # Pass the tokens through the model to get hidden states
    outputs = model(**inputs)

    # outputs[0] contains the hidden states (context vectors) for each token
    hidden_states = outputs[0]

    # Tokenize the phrase separately to know how many tokens it contains
    phrase_tokens = tokenizer.encode("Quantum information theory")

    # Find the starting index of "Quantum information theory" in the tokenized sentence
    start_index = inputs['input_ids'][0].tolist().index(phrase_tokens[0])

    # Extract the context vectors for the phrase "Quantum information theory"
    phrase_context_vectors = hidden_states[0, start_index : start_index + len(phrase_tokens), :]

    # Compute pairwise Euclidean distances and store in a matrix
    distance_matrix = torch.zeros(len(phrase_tokens), len(phrase_tokens))
    for i in range(len(phrase_tokens)):
        for j in range(len(phrase_tokens)):
            distance_matrix[i, j] = euclidean_distance(phrase_context_vectors[i], phrase_context_vectors[j])

    print(f'Distance matrix for "Quantum information theory" in sentence "{sentence}":\n {distance_matrix}\n')

    # Perform PCA to reduce the dimensions to 3
    pca = PCA(n_components=3)
    reduced_vectors = pca.fit_transform(phrase_context_vectors.detach().numpy())
    
    # Create labels for each token
    labels = [tokenizer.decode([token_id]) for token_id in phrase_tokens]
    
    # Create a 3D scatter plot
    fig = go.Figure(data=[go.Scatter3d(
        x=reduced_vectors[:, 0],
        y=reduced_vectors[:, 1],
        z=reduced_vectors[:, 2],
        mode='markers+text',
        text=labels,
        marker=dict(
            size=8,
            color=distance_matrix.mean(dim=1).detach().numpy(), # set color to average distance
            colorscale='Viridis',   # choose a colorscale
            opacity=0.8
        ),
        textposition="bottom center"
    )])

    fig.update_layout(scene=dict(
                    xaxis_title='PC1',
                    yaxis_title='PC2',
                    zaxis_title='PC3'),
                    width=700,
                    margin=dict(r=20, b=10, l=10, t=10))
    fig.show()
```

What are the linguistic implications, and the implications for expressivity of GPT-2? 

When certain tokens consistently appear together in different contexts, it suggests a strong association or semantic relationship between those tokens. By maintaining a consistent representation of the pairwise distances between the context vectors of these tokens, a language model can capture and encode this association. This consistent representation allows the model to recognize the shared characteristics or semantic connections between the tokens, regardless of the specific context in which they appear.

This consistent representation of pairwise distances enables the language model to generalize the relationships between tokens and learn robust representations that capture their shared meaning or contextual dependencies. It allows the model to leverage the knowledge gained from one context to inform its understanding and generation of those tokens in different contexts.

By having a consistent representation, the model can establish a reliable and stable understanding of the associations between tokens, which can benefit downstream tasks such as information retrieval, question-answering, and text generation. It enables the model to make more accurate predictions and generate coherent and contextually appropriate responses.

However, it's important to note that while a consistent representation of pairwise distances can be beneficial for capturing common associations between tokens, it may also limit the model's ability to capture subtle variations or context-specific relationships. Depending on the task and the desired level of granularity, the trade-off between consistency and flexibility needs to be carefully considered when designing and evaluating language models.

```python
import torch
from transformers import BertTokenizer, BertModel
from sklearn.decomposition import PCA
import plotly.graph_objects as go

# Initialize the model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Function to calculate Euclidean distance
def euclidean_distance(a, b):
    return torch.dist(a, b)

# Prompt user to enter two sentences
sentences = [input("Enter first sentence: "), input("Enter second sentence: ")]

# Iterate over sentences
for sentence in sentences:
    # Tokenize the sentence
    inputs = tokenizer(sentence, return_tensors='pt')

    # Pass the tokens through the model to get hidden states
    outputs = model(**inputs)

    # outputs.last_hidden_state contains the hidden states (context vectors) for each token
    hidden_states = outputs.last_hidden_state

    # Tokenize the phrase separately to know how many tokens it contains
    phrase_tokens = tokenizer.encode("Quantum information theory", add_special_tokens=False)

    # Find the starting index of "Quantum information theory" in the tokenized sentence
    input_ids = inputs['input_ids'].tolist()[0]
    start_index = [i for i, val in enumerate(input_ids) if val == phrase_tokens[0]][0]

    # Extract the context vectors for the phrase "Quantum information theory"
    phrase_context_vectors = hidden_states[0, start_index : start_index + len(phrase_tokens), :]

    # Compute pairwise Euclidean distances and store in a matrix
    distance_matrix = torch.zeros(len(phrase_tokens), len(phrase_tokens))
    for i in range(len(phrase_tokens)):
        for j in range(len(phrase_tokens)):
            distance_matrix[i, j] = euclidean_distance(phrase_context_vectors[i], phrase_context_vectors[j])

    print(f'Distance matrix for "Quantum information theory" in sentence "{sentence}":\n {distance_matrix}\n')

    # Perform PCA to reduce the dimensions to 3
    pca = PCA(n_components=3)
    reduced_vectors = pca.fit_transform(phrase_context_vectors.detach().numpy())

    # Create labels for each token
    labels = [tokenizer.decode(token_id) for token_id in phrase_tokens]

    # Create a 3D scatter plot
    fig = go.Figure(data=[go.Scatter3d(
        x=reduced_vectors[:, 0],
        y=reduced_vectors[:, 1],
        z=reduced_vectors[:, 2],
        mode='markers+text',
        text=labels,
        marker=dict(
            size=8,
            color=distance_matrix.mean(dim=1).detach().numpy(),  # set color to average distance
            colorscale='Viridis',  # choose a colorscale
            opacity=0.8
        ),
        textposition="bottom center"
    )])

    fig.update_layout(scene=dict(
        xaxis_title='PC1',
        yaxis_title='PC2',
        zaxis_title='PC3'),
        width=700,
        margin=dict(r=20, b=10, l=10, t=10))
    fig.show()
```

As we can see, the Bert model does not maintain the pairwise distances between the tokens' context vectors, and we get two different distance matrices. The varying pairwise distances observed in BERT's context vectors for the phrase "Quantum information theory" across different sentences indicate that BERT does not encode the tokens as a unique conceptual unit.

BERT's contextualized representations are designed to capture fine-grained contextual information and encode the relationships between tokens based on their specific context. In this case, the varying distances suggest that BERT differentiates the context vectors for each token in the phrase based on the surrounding words and their specific contextual associations.
